## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
query = '''
select distinct(launch_site) from spacex;
'''

ps.sqldf(query, locals())
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
query = '''
select * from spacex
where launch_site like 'CCA%'
limit 5
;'''

ps.sqldf(query, locals())
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
query = '''
select sum(payload_mass__kg_) as total_payload_mass
from spacex
where customer like '%NASA (CRS)%'
;'''

ps.sqldf(query, locals())
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
query = '''
select avg(payload_mass__kg_) as avg_payload_mass
from spacex
where booster_version like '%F9 v1.1%'
;'''

ps.sqldf(query, locals())
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
query = '''
select DATE, Landing__Outcome
from spacex
where Landing__Outcome = 'Success (ground pad)'
order by DATE
limit 1
;'''

ps.sqldf(query, locals())
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
query = '''
select booster_version, landing__outcome, payload_mass__kg_ 
from spacex
where landing__outcome like '%Success (drone ship)%'
and payload_mass__kg_ > 4000
and payload_mass__kg_ < 6000
;'''

ps.sqldf(query, locals())
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
query = '''
select mission_outcome, count(mission_outcome) as total_number
from spacex
group by mission_outcome
;'''

ps.sqldf(query, locals())
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
query = '''
select distinct booster_version, payload_mass__kg_  
from spacex
where payload_mass__kg_ = (select max(payload_mass__kg_) from spacex)
;
'''

ps.sqldf(query, locals())
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
query = '''
select landing__outcome, booster_version, launch_site, DATE
from spacex
where landing__outcome like '%Failure (drone ship)%' and DATE like '%2015%'
;'''

ps.sqldf(query, locals())
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
query = '''
select 
rank() over(order by count(landing__outcome) desc) as Rank,
landing__outcome, count(landing__outcome) as count_of_landing__outcome
from spacex
where DATE > '2010-06-04'
and DATE < '2017-03-20'
group by landing__outcome
order by count_of_landing__outcome desc
;'''

ps.sqldf(query, locals())
```
